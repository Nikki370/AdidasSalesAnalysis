```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Image, display
```

```python
top_retailers = df.groupby('Retailer')['Total Sales'].sum().sort_values(ascending=False)
print(top_retailers)
```

```python
top_categories = df.groupby('Product Category')['Total Sales'].sum().sort_values(ascending=False)
print(top_categories)
```

```python
sales_method = df.groupby('Sales Method')['Total Sales'].sum()
print(sales_method)
```

```python
profit_city = df.groupby('City')['Operating Profit'].sum().sort_values(ascending=False)
print(profit_city)
```

```python
colors = plt.cm.viridis(np.linspace(0, 1, len(top_retailers)))
top_categories.plot(kind='bar', color=colors)
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
product_summary = df.groupby('Product Category').agg({
    'Units Sold': 'sum',
    'Total Sales': 'sum',
    'Operating Profit': 'sum',
    'Operating Margin': 'mean'
}).sort_values(by='Operating Profit', ascending=False).reset_index()
product_summary
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Operating Profit', y='Product Category', data=product_summary, hue=product_summary.index, palette='crest')
plt.title('Operating Profit by Product Category')
plt.xlabel('Profit')
plt.ylabel('Product')
plt.show()
```

```python
retail_summary = df.groupby('Retailer').agg({
    'Total Sales': 'sum',
    'Operating Profit': 'sum',
    'Operating Margin': 'mean'
}).sort_values(by='Operating Profit', ascending=False).reset_index()
retail_summary
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Operating Profit', y='Retailer', data=retail_summary, hue=retail_summary.index, palette='magma')
plt.title('Retailer-wise Operating Profit')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Operating Margin', y='Retailer', data=retail_summary, hue=retail_summary.index, palette='viridis')
plt.title('Retailer-wise Operating Margins')
plt.show()
```

```python
region_perf = df.groupby(['Region', 'Product Category'])['Total Sales'].sum().unstack().fillna(0)
plt.figure(figsize=(10,6))
sns.heatmap(region_perf, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title('Sales by Region & Product Category')
plt.show()
```

```python
monthly_trend = df.groupby('Month')[['Total Sales', 'Operating Profit', 'Units Sold']].sum().reset_index()
monthly_trend
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=monthly_trend, x='Month', y='Total Sales', marker='o', label='Total Sales')
sns.lineplot(data=monthly_trend, x='Month', y='Operating Profit', marker='o', label='Operating Profit')
plt.title('Monthly Sales & Profit Trends')
plt.xticks(range(1,13))
plt.grid(True)
plt.legend()
plt.show()
```

```python
time_series = df.groupby(df['Invoice Date'].dt.to_period('M')).agg({
    'Total Sales': 'sum',
    'Operating Profit': 'sum'
}).reset_index()
time_series['Invoice Date'] = time_series['Invoice Date'].astype(str)
time_series
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Invoice Date', y='Total Sales', data=time_series, label='Sales', marker='o')
sns.lineplot(x='Invoice Date', y='Operating Profit', data=time_series, label='Profit', marker='o')
plt.xticks(rotation=45)
plt.title('Sales and Profit Over Time')
plt.tight_layout()
plt.legend()
plt.show()
```

```python
gender_prod = df.groupby(['Gender Type', 'Product Category'])[['Total Sales', 'Operating Profit']].sum().unstack().fillna(0)
gender_prod.columns = [f'{col[0]} - {col[1]}' for col in gender_prod.columns]
gender_prod
```

```python
gender_prod[['Operating Profit - Apparel', 'Operating Profit - Athletic Footwear', 'Operating Profit - Street Footwear']].plot(
    kind='bar', stacked=True, figsize=(10,6), colormap='tab20', width=0.8
)
plt.title('Operating Profit by Gender and Product Category')
plt.ylabel('Operating Profit')
plt.xticks(rotation=0)
plt.legend(title='Product Category')
plt.tight_layout()
plt.show()
```

```python
gender_prod[['Total Sales - Apparel', 'Total Sales - Athletic Footwear', 'Total Sales - Street Footwear']].plot(
    kind='bar', stacked=True, figsize=(10,6), colormap='tab20', width=0.8
)
plt.title('Total Sales by Gender and Product Category')
plt.ylabel('Sales')
plt.xticks(rotation=0)
plt.legend(title='Product Category')
plt.tight_layout()
plt.show()
```

```python
city_perf = df.groupby('City')[['Operating Profit', 'Total Sales']].sum().sort_values(by='Operating Profit', ascending=False).head(10).reset_index()
city_perf
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Operating Profit', y='City', data=city_perf, hue=city_perf.index, palette='coolwarm')
plt.title('Top 10 Cities by Operating Profit')
plt.show()
```

```python
sales_summary = df.groupby('Sales Method').agg({
    'Total Sales': 'sum',
    'Operating Profit': 'sum',
    'Operating Margin': 'mean'
}).reset_index()
sales_summary
```

```python
method_dist = df['Sales Method'].value_counts()
colors = sns.color_palette("pastel")
plt.pie(method_dist, labels=method_dist.index, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title('Sales by Method')
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(14,6))
sns.scatterplot(data=df, x='Units Sold', y='Total Sales', hue='Product Category', ax=ax[0], palette='deep')
ax[0].set_title('Units Sold vs Total Sales')
sns.scatterplot(data=df, x='Price per Unit', y='Total Sales', hue='Product Category', ax=ax[1], palette='dark')
ax[1].set_title('Price vs Total Sales')
plt.tight_layout()
plt.show()
```
